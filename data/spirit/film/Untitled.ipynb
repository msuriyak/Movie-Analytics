```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('spirit_film_1999.csv')
```

```python
df
```

```python
res = []
for year in df['Year'].unique():
    temp = df[df['Year'] == year]
    if type(temp) == pd.core.series.Series:
        res += [1]
    else: 
        res += [1] + [0 for i in range(len(temp) -1 )]
df['Spirit best film winner'] = res
df['Spirit best film nominee'] = 1
```

```python
df2 = pd.read_csv('films.csv')
```

```python
df_final = pd.concat([df, df2]).reset_index()
```

```python
df_final = df_final[df_final.columns[1:].tolist()]
```

```python
df_final.to_csv('films_spirit.csv', index=False)
```
